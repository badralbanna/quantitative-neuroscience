## Setup

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# For retina displays only 
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

## `Hello World`

```python
X = np.random.rand(100)
X2 = np.random.normal(size=100)
Y = np.random.rand(100)
```

```python
plt.scatter(X2, Y)
```

```python
X2 = np.linspace(-10, 10, 1000)
Y2 = np.sin(X2)
```

```python
plt.plot(X2, Y2)
```

## Figures, subplots, multiple plots 

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(X2, Y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 6))
ax1.scatter(X, Y)
ax2.plot(X2, Y2)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 6))
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 6), sharey=True)
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2)
```

```python
fig, axs = plt.subplots(nrows=10, ncols=4, figsize=(10, 6))
axs[0,0].scatter(X, Y)
axs[0,1].plot(X2, Y2)
```

```python
Y3 = np.cos(X2)

fig, axs = plt.subplots(ncols=2, figsize=(10, 6), sharey=True)
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2)
axs[1].plot(X2, Y3)
fig.tight_layout()
```

## Annotating plots

```python
# using a legend
fig, axs = plt.subplots(ncols=2, figsize=(10, 6), sharey=True)
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2, label="$\sin(\chi)$")
axs[1].plot(X2, Y3, label="$\cos(\chi)$")
axs[1].set_ylim(-3, 1)
fig.tight_layout()
fig.legend(loc='lower right')
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 6), sharey=True)
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2, label="$\sin(\chi)$")
axs[1].plot(X2, Y3, label="$\cos(\chi)$")
axs[1].set_ylim(-3, 1)
axs[1].text()
fig.tight_layout()
```

## Customizing plot styles

### Data

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 6), sharey=True)
axs[0].scatter(X, Y)
axs[1].plot(X2, Y2, label="$\sin(\chi)$")
axs[1].plot(X2, Y3, label="$\cos(\chi)$")
axs[1].set_ylim(-3, 1)
axs[1].set_ylabel("intensity")
axs[1].set_xlabel("$\chi$")
ax1 = axs[1]
ax1.set_
fig.tight_layout()
```

### Globally

```python
plt.rcParams['legend.fontsize'] = 'small'
plt.rcParams['axes.labelsize'] = 'medium'
plt.rcParams['axes.formatter.limits'] = [-2, 3]
plt.rcParams['axes.formatter.use_mathtext'] = True
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['figure.dpi'] = 150
```
